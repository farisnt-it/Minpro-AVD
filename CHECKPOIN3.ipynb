## Outliers

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

# Construct Data


"Revenue per Unit"

```python
df['Revenue per Unit'] = df['Revenue ($)'] / df['Units Sold']
```

5G vs Non-5G

```python
df['5G vs Non-5G'] = df['5G Capability'].apply(lambda x: '5G' if x == 'Yes' else 'Non-5G')
```

Market Share Growth

```python
df['Market Share Growth'] = df.groupby('Product Model')['Market Share (%)'].diff()
```

Regional 5G Adoption Rate

```python
df['Regional 5G Adoption Rate'] = pd.cut(
    df['Regional 5G Coverage (%)'],
    bins=[0, 40, 70, 100],
    labels=['Rendah', 'Sedang', 'Tinggi'])
```

Preference for 5G (%)

```python
df['Preference Category'] = pd.cut(
    df['Preference for 5G (%)'],
    bins=[0, 40, 70, 100],
    labels=['Rendah', 'Sedang', 'Tinggi']
)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Analisis Tren Penjualan Produk Samsung Galaxy dari Tahun 2019 hingga 2024 dengan Fokus pada Produk yang Mendukung Teknologi 5G

```python
# Visualisasi Revenue per Unit untuk 5G vs. Non-5G
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='5G vs Non-5G', y='Revenue per Unit')
plt.title('Revenue per Unit untuk Produk 5G vs. Non-5G')
plt.xlabel('Kategori Produk')
plt.ylabel('Revenue per Unit ($)')
plt.show()
```

Insight:
1. Revenue per Unit 5G: Produk dengan kemampuan 5G cenderung memiliki revenue per unit yang lebih tinggi dibandingkan produk non-5G. Ini menunjukkan bahwa konsumen bersedia membayar lebih untuk produk dengan teknologi 5G.

2. *Revenue* per Unit Non-5G: Produk non-5G memiliki revenue per unit yang lebih rendah, yang mungkin disebabkan oleh harga yang lebih terjangkau atau kurangnya fitur premium.

```python
# Visualisasi Market Share Growth untuk Produk 5G vs. Non-5G
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='Year', y='Market Share Growth', hue='5G vs Non-5G', ci=None, estimator='mean')
plt.title('Pertumbuhan Pangsa Pasar untuk Produk 5G vs. Non-5G')
plt.xlabel('Tahun')
plt.ylabel('Pertumbuhan Pangsa Pasar (%)')
plt.legend(title='Kategori Produk')
plt.show()
```

Insight:
1. Tren Penjualan 5G: Penjualan produk dengan kemampuan 5G menunjukkan peningkatan yang signifikan dari tahun ke tahun, terutama setelah tahun 2020. Ini menunjukkan bahwa adopsi teknologi 5G semakin meningkat di kalangan konsumen.

2. Tren Penjualan Non-5G: Penjualan produk tanpa kemampuan 5G cenderung stagnan atau menurun, terutama setelah tahun 2021. Hal ini menunjukkan bahwa konsumen lebih memilih produk dengan teknologi 5G.

```python
# Visualisasi Penjualan per Region berdasarkan Tingkat Adopsi 5G
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Region', y='Units Sold', hue='Regional 5G Adoption Rate', ci=None, estimator=sum)
plt.title('Penjualan per Region berdasarkan Tingkat Adopsi 5G')
plt.xlabel('Region')
plt.ylabel('Total Unit Terjual')
plt.legend(title='Tingkat Adopsi 5G')
plt.show()
```

Insight:
1. Region dengan Adopsi Tinggi: Region seperti North America dan Asia-Pacific memiliki tingkat adopsi 5G yang tinggi, yang sejalan dengan penjualan produk 5G yang tinggi.

2. Region dengan Adopsi Rendah: Region seperti Middle East & Africa dan Latin America memiliki tingkat adopsi 5G yang rendah, yang memengaruhi penjualan produk 5G di region tersebut.

```python
# Visualisasi Penjualan berdasarkan Preferensi 5G
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Preference Category', y='Units Sold', hue='5G vs Non-5G', ci=None, estimator=sum)
plt.title('Penjualan berdasarkan Preferensi Konsumen terhadap 5G')
plt.xlabel('Preferensi 5G')
plt.ylabel('Total Unit Terjual')
plt.legend(title='Kategori Produk')
plt.show()
```

Insight:
1. Preferensi Tinggi: Di region dengan preferensi 5G tinggi, penjualan produk 5G sangat dominan.

2. Preferensi Rendah: Di region dengan preferensi 5G rendah, penjualan produk non-5G masih mendominasi.
